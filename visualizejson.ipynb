```python
import sys
sys.path += ['../..']

import os

import tarfile
import json

import matplotlib.pyplot as plt

import string
```

```python
PATH_TO_CWL_DATA = '/home/vagrant/work/week6/cwl-data/data/structured/'
EVENT = 'structured-2018-01-14-neworleans/'
MATCH = 'structured-1515788270-56925b97-0ea7-5d64-b76b-829c30867fee.json'

PATH_TO_WW2_MAPS = '/home/vagrant/work/week6/cwl-data/maps/ww2/'

MAP_IMAGES = {
    'Aachen': 'aachen.png',
    'Ardennes Forest': 'ardennes_forest.png',
    'Flak Tower': 'flaktower.png',
    'Gibraltar': 'gibraltar.png',
    'London Docks': 'london_docks.png',
    'Sainte Marie du Mont': 'sainte_marie_du_mont.png' ,
    'USS Texas': 'uss_texas.png',
    'Valkyrie': 'valkyrie.png'
}
```

```python
import re
pattern = r'(\w+-2018-\d+-\d+-\w+)'

EVENTS_DIR = os.listdir('/home/vagrant/work/week6/cwl-data/data/structured/')
EVENTS = []
for file in EVENTS_DIR:
    if not file.endswith(".tar.gz") and len(file) > 15 and re.match(pattern, file):
        EVENTS.append(file)
```

```python
for x_event in EVENTS:
    os.mkdir('/home/vagrant/work/week6/maps_hw/' + x_event)
```

```python
# load match data and plot match events
for x_event in EVENTS:
    for y_match in os.listdir('/home/vagrant/work/week6/cwl-data/data/structured/' + x_event):
        with open('{}{}/{}'.format(PATH_TO_CWL_DATA, x_event, y_match), 'r') as match_file:
            # get match data
            match = json.load(match_file)

            kill_pos, death_pos, spawn_pos = {'x': [], 'y': []}, {'x': [], 'y': []}, {'x': [], 'y': []}

            # get pos data
            events = match['events']
            for evt in events:
                if evt['type'] == 'spawn':
                    spawn_pos['x'].append( evt['data']['pos']['x'] )
                    spawn_pos['y'].append( evt['data']['pos']['y'] )
                if evt['type'] == 'death':
                    kill_pos['x'].append( evt['data']['attacker']['pos']['x'] )
                    kill_pos['y'].append( evt['data']['attacker']['pos']['y'] )
                    death_pos['x'].append( evt['data']['pos']['x'] )
                    death_pos['y'].append( evt['data']['pos']['y'] )

            # get map image
            map_file_name = PATH_TO_WW2_MAPS + MAP_IMAGES[match['map']]
            map_image = plt.imread(map_file_name)

            # create figure
            fig, ax = plt.subplots(1,1)
            fig.set_figheight(15)
            fig.set_figwidth(15)
            # plot map
            ax.imshow(map_image)
            ax.set_facecolor((210/255.0, 210/255.0, 210/255.0))
            # plot spawns amd deaths
            ax.scatter(spawn_pos['x'], spawn_pos['y'], color='blue')
            ax.scatter(death_pos['x'], death_pos['y'], color='red')
            plt.savefig('/home/vagrant/work/week6/maps_hw/' + x_event +'/' +  y_match.rstrip('.json') + '.png')
            plt.close()
```
